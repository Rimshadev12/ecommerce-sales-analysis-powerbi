```python
import pandas  as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')


df = pd.read_csv("Ecommerce_Dataset.csv")
df.head()
```

```python
df.shape
```

```python
df = df.dropna(axis=1, how='all')
```

```python
df = df.dropna(axis=0, how='all')
```

```python
df = df.reset_index(drop=True)
```

```python
df.shape
```

```python
# Strip whitespace, lowercase, replace spaces and dashes with underscore
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_").str.replace("-", "_")

print(df.columns)
```

```python
df = df.dropna(how='all')
df = df.dropna(axis=1, how='all')
```

```python
df.info()
```

```python
df['item_id'] = df['item_id'].astype('Int64')
df['customer_id'] = df['customer_id'].astype('Int64')
```

```python
df['created_at'] = pd.to_datetime(df['created_at'])
df['working_date'] = pd.to_datetime(df['working_date'])
```

```python
df['mv'] = df['mv'].astype(str)
```

```python
import numpy as np

def clean_mv(value):
    try:
        # Convert to string first
        s = str(value).replace(',', '').strip()
        # If empty or invalid, return NaN
        if s in ('', '-', 'nan', 'None'):
            return np.nan
        # Convert to float
        return float(s)
    except:
        return np.nan

# Apply to the whole column
df['mv'] = df['mv'].apply(clean_mv)
```

```python
# For example, fill invalid entries with 0
df['mv'] = df['mv'].fillna(0)
```

```python
df['price'] = df['price'].astype(float)
df['qty_ordered'] = df['qty_ordered'].astype(float)
df['grand_total'] = df['grand_total'].astype(float)
df['discount_amount'] = df['discount_amount'].astype(float)
df['year'] = df['year'].astype(int)
df['month'] = df['month'].astype(int)
```

```python
print(df.isna().sum())
```

```python
# Remove rows where critical data is missing
df = df.dropna(subset=['status', 'sku', 'customer_id'])

# Fill optional / categorical columns
df['category_name_1'] = df['category_name_1'].fillna('Unknown')

# sales_commission_code stays as is (NaN)
# customer_since can stay as NaN
```

```python
# IDs
df['item_id'] = df['item_id'].astype('Int64')
df['customer_id'] = df['customer_id'].astype('Int64')

# Dates
df['created_at'] = pd.to_datetime(df['created_at'])
df['working_date'] = pd.to_datetime(df['working_date'])
df['customer_since'] = pd.to_datetime(df['customer_since'], errors='coerce')

# Numeric
num_cols = ['price','qty_ordered','grand_total','discount_amount','mv']
df[num_cols] = df[num_cols].astype(float)

# Integer columns
df['year'] = df['year'].astype(int)
df['month'] = df['month'].astype(int)

# Strip strings
str_cols = ['status','sku','category_name_1','sales_commission_code',
            'payment_method','bi_status','m_y','fy']
for col in str_cols:
    df[col] = df[col].str.strip()
```

```python
df = df.reset_index(drop=True)
```

```python
print(df.isna().sum())
```

```python
df.to_csv("ecommerce_clean.csv", index=False)
```
